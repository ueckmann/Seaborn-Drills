# Thinkful Unit 1 Lesson 3.3 Drill: Present data in multiple ways

```python
# Import numpy, pandas, matplotlib, and seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
#import our dataframe
df = pd.read_excel('Pokemon.xlsx')
df.rename(columns={'#':'Number'}, inplace=True)
df.set_index('Number')
df.head(n=10)
```

## 1. Choose one variable and plot that variable four different ways.

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.boxplot(data = df, y = 'HP')
plt.title("Boxplot of HP")

plt.subplot(2,2,2)
sns.lineplot(data = df, x = df.index, y ='HP')
plt.xlabel('Pokemon Number')
plt.title('Line Graph of HP')

plt.subplot(2,2,3)
sns.distplot(df['HP'], kde= False)
plt.title('Histogram of HP')

plt.subplot(2,2,4)
sns.distplot(df['HP'], hist= False)
plt.title('Density Plot of HP')

plt.tight_layout
plt.show()
```

## 2. Choose two continuous variables, and plot them three different ways.

```python
plt.figure(figsize=(15,15))

plt.subplot(3,1,1)
sns.boxplot(data = df,x = 'Attack', y = 'Defense')
plt.title("Boxplot of Attack and Defense")

plt.subplot(3,1,2)
sns.scatterplot(data = df, x = 'Attack', y = 'Defense')
plt.title('Scatter Plot of Attack and Defense')

plt.subplot(3,1,3)
sns.swarmplot(data = df, x = 'Attack', y = 'Defense')
plt.title('Swarm Plot of Attack and Defense')

plt.tight_layout
plt.show()
```

## 3. Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
plt.figure(figsize=(15,30))

plt.subplot(6,1,1)
sns.boxplot(data = df,x = 'Type 1', y = 'Attack')
plt.title("Box Plot of Pokemon Type Attack Strength")
plt.xlabel('Pokemon Type')

plt.subplot(6,1,2)
sns.scatterplot(data = df, x = 'Type 1', y = 'Attack')
plt.title('Scatter Plot of Pokemon Type Attack Strength')

plt.subplot(6,1,3)
sns.lineplot(data = df, x = 'Type 1', y = 'Attack')
plt.title('Line Plot of Pokemon Type Attack Strength')

plt.subplot(6,1,4)
sns.stripplot(data = df,x = 'Type 1', y = 'Attack')
plt.title("Strip Plot of Pokemon Type Attack Strength")
plt.xlabel('Pokemon Type')

plt.subplot(6,1,5)
sns.swarmplot(data = df, x = 'Type 1', y = 'Attack')
plt.title('Swarm Plot of Pokemon Type Attack Strength')

plt.subplot(6,1,6)
sns.violinplot(data = df, x = 'Type 1', y = 'Attack')
plt.title('Violin Plot of Pokemon Type Attack Strength')


plt.tight_layout
plt.show()
```
